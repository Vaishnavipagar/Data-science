```python
plt.scatter(x,y,c='g')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
plt.show()
```

```python
## line plot
plt.plot(x, y, 'ro', linestyle='dashed', linewidth=2, markersize=12)
plt.xlabel('X axis')
plt.ylabel('y axis')
plt.title('2d diagram')
plt.show()  # Optional, if you want to display the plot
```

```python
##creating subplots
## Assignment--plot x title y title
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(y,x,'g--')
plt.subplot(2,2,3)
plt.plot(x,y,'b--')
plt.subplot(2,2,4)
plt.plot(x,y,'b*')
plt.show()
```

```python
## bar plotss
x1=[2,3,4,5]
y1=[11,16,19,20]

x2=[3,9,11,14]
y2=[6,15,7,9]
plt.bar(x1,y1,color='g')
plt.bar(x2,y2)
plt.show()
```

```python
## Histograms

a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a,bins=5)
plt.show()
```

```python
## Boxplot ussing Matplotlib
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(data)
plt.show()
```

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array(["A", "B", "C", "D"])  # Categories (y-axis in horizontal bar chart)
y = np.array([3, 8, 1, 10])        # Values (x-axis in horizontal bar chart)

plt.barh(x, y)  # Creates horizontal bar chart
plt.show()      # Displays the chart
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show() 
```
